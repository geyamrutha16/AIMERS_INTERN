```python
import telebot
import requests
import json
import os
import openai
from openai import OpenAI

# Replace with your API's
OPENWEATHER_API_KEY = ""
TELEGRAM_API_KEY = ""
OPENAI_API_KEY = ""
client = OpenAI(api_key=OPENAI_API_KEY)

bot = telebot.TeleBot(TELEGRAM_API_KEY)

@bot.message_handler(commands=['start', 'help'])
def send_welcome(message):
    bot.reply_to(message, "Hello..! How can I help you .. ?\n 1. Weather Info \n 2. Other")

@bot.message_handler(commands=["3"])
def gen_information(message):
    bot.reply_to(message, "Enter your question :")
    bot.register_next_step_handler(message, get_info)

def get_info(message):
    try :
      print(message)
      completion = client.chat.completions.create(
      model="gpt-3.5-turbo",
      messages=[
        {"role": "user", "content": message.text},
        ]
      )
      bot.reply_to(message, completion.choices[0].message.content)
    except Exception as e:
        print(f"An error occurred: {e}")
        bot.reply_to(message, "Sorry, I couldn't process your request.")
    bot.reply_to(message, "How can I help you .. ?\n 1. Weather Info \n 2. Other")

@bot.message_handler(commands=["2"])
def get_city_for_weather(message):
    bot.reply_to(message, "Please enter a city name for weather information:")
    bot.register_next_step_handler(message, get_weather)

def get_weather(message):
    city_name = message.text
    response = requests.get(f'http://api.openweathermap.org/data/2.5/weather?q={city_name}&appid={OPENWEATHER_API_KEY}')
    data = json.loads(response.text)
    if data['cod'] == 200:
        temp = data['main']['temp'] - 273.15  # Convert from Kelvin to Celsius
        bot.reply_to(message, f'The current temperature in {city_name} is {temp:.2f}°C.')
    else:
        bot.reply_to(message, 'Sorry, I could not find that city.')
    bot.reply_to(message, "How can I help you .. ?\n 1. Weather Info \n 2. Other")

@bot.message_handler(func=lambda message: True)
def echo_all(message):
    if(message.text == "1"):
      get_city_for_weather(message)
    if(message.text == "2"):
      gen_information(message)
    if(message.text != "1" and message.text != "2"):
      bot.reply_to(message, "Sorry, I don't understand that command.")
      bot.reply_to(message, "How can I help you .. ?\n 1. Weather Info \n 2. Other")


bot.polling()
```
